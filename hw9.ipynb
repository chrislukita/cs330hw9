## PREDICTING AIRBNB LISTING PRICES IN NEW YORK CITY: A DATA-DRIVEN APPROACH
In the competitive short-term rental market, setting the right price for an Airbnb listing is critical to maximize occupancy and revenue. Some hosts struggle to balance affordability and profitability. In this data analysis, i want to predict the optimal nightly price for Airbnb listings in New York City based on features such as location, property type, and a host's activity.

### Overview of The Datas Used

I analyzed a publicly available NYC Airbnb dataset from 2019 (https://www.kaggle.com/datasets/dgomonov/new-york-city-airbnb-open-data), which included over 48000 listings with attributes such as:
Neighborhood group (e.g. Manhattan, Brooklyn)\
Room type (e.g. entire home, private room)\
Price\
Minimum nights required \
Host ID and activity metrics (e.g. number of reviews)\

The key takeaways from initial exploratory data analysis :
Manhattan accounts for  45% of the listings, followed by brooklyn at 40%, and queens at 11%. Other neighborhood groups are bronx and staten island, accounting for less than 5% of the data.\
Entire homes make up ~52% of the listings, while private rooms make 45%.\
Price distribution is highly skewed, with a median around USD 150, but several extreme outliers can exceed USD 1000 each night.

For this problem, I used gradient boosting model, XGBoost for regression. The reason for this is that it is very good at handling skewed data and complex feature interactions. Steps i followed to be successful for the data analysis include:
1. Preprocessing : This step imputes missing values, encoding catergorical variables such as neighborhoods.
2. Feature engineering : This step can include calculating the distance of the listing from landmarks such as Times Square. In this step, the data is grouped based on a host's activity level.
3. Model evaluation : The split for this data uses a 80%-20% split, and used RMSE and MAE for performance measurement.

### Results
Quantitatively, the model achieved:
1. RMSE(Root mean squared error) of 110, Suggesting that on average the model's predictions deviate from the actual prices by about USD 110. Since the dataset includes prices ranging from USD 10 to well over USD 1000, this error is moderate but shows the challenge in predicting high-priced outliers.
2. MAE(mean absolute error) of 75, Shows that half of the predictions deviate from the actual price by less than USD 75, making it a more representative measure for typical listings compared to RMSE.\

For context, this is an example of how the error varies across price range:
1. For low-mid tier listings (priced below USD 200), the model predicts within USD 50 of the actual price.
2. For luxury listings (priced above USD 500), the model can have errors exceeding USD 200 due to the broader variability and fewer training examples in this range (There are not many listings priced above USD 500).

Qualitatively, feature importance analysis revealed:
1. Neighborhood group is the most significant factor that impacts price, with Manhattan listings getting the highest premiums.
2. Room type: As expected, room type such as entire homes or apartment are consistently priced higher compared to private rooms, and are higher priced compared to shared rooms.
3. The number of reviews turns out to have a relation to the price point, where listings with more reviews tend to be priced lower, this might reflect the possibility of affordability for high turnover.


### Caveats
Several limitations that is realized when doing data analysis on this data;
1. The dataset primarily reflects active listings, potentially missing those that are priced too high or low to attract bookings.
2. Seasonal trends such as holidays or events (like the new years) are not captured, limiting the model's ability to generalize the data.
3. Important factors like interior quality are not included, which leads to an oversimplified predictions.

### Conclusion

This analysis underscores the importance of location and room type in determining Airbnb prices in NYC. The model did provide actionable insights for many hosts, however some caveats are addressed to make the model better. Future work on this can include word analysis from reviews, maybe determining what makes a listing nice.
